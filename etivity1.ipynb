## Pivot Table
Pivot table for people with personal loans based on their education level. 


*   Primary - 11%
*   Secondary - 14%
*   Tertiary - 11%





```python
pivot_table_LC = df.pivot_table(values='loan',
                                index='education',
                                aggfunc=lambda x: x.map({'yes':1, 'no':0}).mean())
```

```python
print(pivot_table_LC)
```

```python
# Plot pivot table
pivot_table_LC.plot(kind='bar')
plt.xlabel('Education Level')
plt.ylabel('Probability of personal loan ')
plt.title('Probability of personal loan based on their education')
plt.legend().set_visible(False)
```

## Stacked Chart 

This stacked chart shows the correlation between if a person has a personal loan and a housing loan. This shows that more people dont have any sort of loan and fewer again have both a personal and house loan 

```python
# Use crosstab function to create a stacked chart of personal loan and housing loan
crosstab_CL = pd.crosstab(df['housing'], df['loan'])
crosstab_CL.plot(kind='bar', stacked=True, color=['red', 'green'])
plt.ylabel('Number of Applicants')
plt.show()
```

## Advanced Stacked Chart
In this stacked chart, I have added an education tab to examine the relationship with the different levels of education and people having loans. 

```python
# Added education to the stackchart to examine the relationship with education and if they had any loan.
df = df.replace({'housing': { 1 :  'yes', 0 : 'no'}}) 
crosstab_CGL = pd.crosstab([df['housing'], df['loan']], df['education'])
crosstab_CGL.plot(kind='bar', stacked=True, color=['red', 'orange', 'green'], grid=False)
plt.ylabel('Number of Applicants')
plt.show()
```

## Conclusion

It is quite evident that the majority of people do not have loans. 
The number of those that have personal loans and dont have housing loans is extremely low in comparison to those who have housing loans and dont have personal loans.
Less than 200 people have both types of loans and over 800 have no loans at all. 
The higher the person's balance is the more likely they are to not have a loan. 
Those that have tertiary level education are highly unlikely to have any form of loans or a personal loan.
There is an even spread on those with primary and secondary education and no loans with those with a housing loan only.
